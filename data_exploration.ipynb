Looks like a really good group of genres would be 'Rock', 'Pop', 'Hip-Hop' and 'Country'

```python
genre_and_lyrics = pd.read_csv(r'datasets/genre_and_lyrics.csv')
genre_and_lyrics = (genre_and_lyrics
                    .loc[genre_and_lyrics['genre']
                    .isin(['Rock', 'Pop', 'Hip-Hop', 'Country'])])
```

```python
def top_n_words_in_genre(n, genre, dataframe):
    genre_df = dataframe[dataframe['genre'] == genre]
    genre_df = genre_df.groupby('word').count().sort_values('count', ascending=False)
    return genre_df[:n]
```

```python
nouns = genre_and_lyrics[genre_and_lyrics['pos'].str.startswith('N', na=False)]
verbs = genre_and_lyrics[genre_and_lyrics['pos'].str.startswith('V', na=False)]
adjectives = genre_and_lyrics[genre_and_lyrics['pos'].str.startswith('J', na=False)]
```

# Comparison

Would be really interesting to chart the top three words of each genre against each other, but first off, we'll need to convert the straight counts into a proportion of total words.

```python
top_nouns = []
for genre in ['Rock', 'Pop', 'Hip-Hop', 'Country']:
    genre_words = top_n_words_in_genre(10, genre, nouns).index.to_list()
    top_nouns.extend(genre_words)

top_nouns = list(set(top_nouns))
```

```python
noun_count = {}
for genre in ['Rock', 'Pop', 'Hip-Hop', 'Country']:
    noun_count[genre] = sum(nouns[nouns['genre'] == genre]['count'])
```

```python
select_nouns = nouns[nouns['word'].isin(top_nouns)].groupby(['genre', 'word'])['count'].sum()
select_nouns = select_nouns.reset_index()
select_nouns['word_proportion'] = select_nouns.apply(
    lambda row: row['count'] / noun_count[row['genre']], axis=1)
```

```python
fig, bar = plt.subplots(figsize = (11,7))
bar = sns.barplot(x="word", y="word_proportion", hue="genre", data=select_nouns)
bar.set_xticklabels(labels=bar.get_xticklabels(), rotation=45)
```

```python
noun_group = ['la', 'yo', 'love', 'night', 'day', 'feel', 'time', 'shit']
small_noun_group = select_nouns[select_nouns['word'].isin(noun_group)]
```

```python
fig, bar = plt.subplots(figsize = (11,7))
bar = sns.barplot(x="word", y="word_proportion", hue="genre", data=small_noun_group)
bar.set_xticklabels(labels=bar.get_xticklabels(), rotation=45)
bar.set_xlabel('Noun')
bar.set_ylabel('Percentage of All Nouns Used')
```

```python
fig, bar = plt.subplots(figsize = (11,10))
bar = sns.barplot(
    x="genre",
    y="word_proportion",
    data=small_noun_group[small_noun_group['word'] == 'love'],
)
bar.set_xticklabels(labels=bar.get_xticklabels())
bar.set_xlabel('')
bar.set_ylabel('Percentage of Nouns that Are "Love"')
```

```python
fig, bar = plt.subplots(figsize = (11,12))
bar = sns.barplot(
    x="genre",
    y="word_proportion",
    data=small_noun_group[small_noun_group['word'] == 'shit'],
)
bar.set_xticklabels(labels=bar.get_xticklabels())
```

```python
top_verbs = []
for genre in ['Rock', 'Pop', 'Hip-Hop', 'Country']:
    genre_words = top_n_words_in_genre(10, genre, verbs).index.to_list()
    top_verbs.extend(genre_words)

top_verbs = list(set(top_verbs))
```

```python
verb_count = {}
for genre in ['Rock', 'Pop', 'Hip-Hop', 'Country']:
    verb_count[genre] = sum(verbs[verbs['genre'] == genre]['count'])
```

```python
select_verbs = verbs[verbs['word'].isin(top_verbs)].groupby(['genre', 'word'])['count'].sum()
select_verbs = select_verbs.reset_index()
select_verbs['word_proportion'] = select_verbs.apply(
    lambda row: row['count'] / verb_count[row['genre']], axis=1)
```

```python
fig, bar = plt.subplots(figsize = (11,7))
bar = sns.barplot(x="word", y="word_proportion", hue="genre", data=select_verbs)
bar.set_xticklabels(labels=bar.get_xticklabels(), rotation=45)
```

```python
verb_group = ['am', 'come', 'get', 'got', 'have', 'know', 'see']
small_verb_group = select_verbs[select_verbs['word'].isin(verb_group)]
```

```python
fig, bar = plt.subplots(figsize = (11,7))
bar = sns.barplot(x="word", y="word_proportion", hue="genre", data=small_verb_group)
bar.set_xticklabels(labels=bar.get_xticklabels(), rotation=45)
bar.set_xlabel('Verb')
bar.set_ylabel('Percentage of All Verbs Used')
```
